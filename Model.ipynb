```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Cleaned_data.csv')
```

```python
data
```

```python
# X y split

X = data.drop(['SalePrice'], axis=1)
y = data['SalePrice']
```

```python
# apply PCA on X

from sklearn.decomposition import PCA
pca = PCA(n_components=35)
X_pca = pca.fit_transform(X)
X_pca
```

```python
X.shape
```

```python
X_pca.shape
```

```python
# Retrieve the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Calculate the cumulative explained variance
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Print the explained variance by each component and the cumulative explained variance
for i, (var_ratio, cum_var) in enumerate(zip(explained_variance_ratio, cumulative_explained_variance)):
    print(f"Principal Component {i+1}: {var_ratio:.2%} variance explained, {cum_var:.2%} cumulative variance explained")
```

```python
# PCA with 15 components
pca = PCA(n_components=15)
X_pca = pca.fit_transform(X)
X_pca
```

```python
# train Linear Regression model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
model_linear = LinearRegression()

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
model_linear.fit(X_train, y_train)

print('Train score:', model_linear.score(X_train, y_train))
print('Test score:', model_linear.score(X_test, y_test))
print('R2 score:', r2_score(y_test, model_linear.predict(X_test)))
print('MSE:', mean_squared_error(y_test, model_linear.predict(X_test)))
```

```python
# Train Random Forest model
from sklearn.ensemble import RandomForestRegressor
model_rf = RandomForestRegressor()
model_rf.fit(X_train, y_train)

print('Train score:', model_rf.score(X_train, y_train))
print('Test score:', model_rf.score(X_test, y_test))
print('R2 score:', r2_score(y_test, model_rf.predict(X_test)))
print('MSE:', mean_squared_error(y_test, model_rf.predict(X_test)))
```

```python
# Train KNN model
from sklearn.neighbors import KNeighborsRegressor
model_knn = KNeighborsRegressor()
model_knn.fit(X_train, y_train)

print('Train score:', model_knn.score(X_train, y_train))
print('Test score:', model_knn.score(X_test, y_test))
print('R2 score:', r2_score(y_test, model_knn.predict(X_test)))
print('MSE:', mean_squared_error(y_test, model_knn.predict(X_test)))
```
